```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('auto-mpg.csv')
df
df.head()
```

```python
#BOX PLOT DIAGRAM
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title('Box plot diagram for entire auto-mpg dataset')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['mpg'])
plt.title('Box plot diagram for mpg attribute')
plt.xlabel('mpg')
plt.show()
```

```python

plt.figure(figsize=(8,5))
sns.boxplot(y=df['mpg'],orient='v')
plt.title('Box plot diagram for mpg attribute')
plt.ylabel('mpg')
plt.show()
```

```python
q1=df['mpg'].quantile(0.25)
q2=df['mpg'].quantile(0.50)
q3=df['mpg'].quantile(0.75)
min_value=df['mpg'].min()
max_value=df['mpg'].max()


plt.figure(figsize=(8,5))
sns.boxplot(y=df['mpg'],orient='v',showfliers=False)
plt.title('Box plot diagram for mpg attribute')
plt.ylabel('mpg')


plt.text(0.85, q1, f'Q1: {q1:.2f}',ha='left', va='center',fontweight='bold')
plt.text(0.85, q2, f'Q2 (median): {q2:.2f}',ha='left', va='center',fontweight='bold')
plt.text(0.85, q3, f'Q3: {q3:.2f}',ha='left', va='center',fontweight='bold')
plt.text(0.85, min_value, f'Min: {min_value:.2f}',ha='left', va='center',fontweight='bold')
plt.text(0.85, max_value, f'Max: {max_value:.2f}',ha='left', va='center',fontweight='bold')

plt.show()
```
