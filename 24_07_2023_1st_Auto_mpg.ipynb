```python
num_data=data.select_dtypes(include=[np.number])
num_data
num_data.head()
```

```python
mean_data=num_data.mean()
median_data=num_data.median()
mode_data=num_data.mode().iloc[0]
variance_data=num_data.var()
std_data=num_data.std()
```

```python
dataframe=pd.DataFrame({
    'Mean':mean_data,
    'Median':median_data,
    'Mode':mode_data,
    'Variance':variance_data,
    'Standard-deviation':std_data
})
dataframe
```
