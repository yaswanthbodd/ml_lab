```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
#load the iris data
data=load_iris()
data
```

```python
x = data.data #independent variables
y = data.target #dependent variables
x
```

```python
y
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(x)
X
```

```python
mean_vectors = []
for c1 in np.unique(y):
  mean_vectors.append(np.mean(X[y==c1],axis=0))
print(mean_vectors)
```

```python
within_class_scatter = np.zeros((X.shape[1],X.shape[1]))
between_class_scatter = np.zeros((X.shape[1],X.shape[1]))
overall_mean = np.mean(X,axis=0)

for c1,mv in zip(np.unique(y),mean_vectors):
  class_scatter = np.cov(X[y == c1].T)
  within_class_scatter += class_scatter
  between_class_scatter += len(X[y == c1]) * ((mv-overall_mean).reshape(-1,1) @ (mv-overall_mean).reshape(1,-1))
```

```python
#compute the eign values and eign vectors

evalues,evectors = np.linalg.eig(np.linalg.inv(within_class_scatter) @ between_class_scatter)
```

```python
evalues ,  evectors
```

```python
#sort the eigenvalues and eigenvectors in descending order

e_pairs = [(np.abs(evalues[i]),evectors[:,i]) for i in range(len(evalues))]
e_pairs.sort(key=lambda X: X[0],reverse=True)
```

```python
e_pairs
```

```python
k = 2
W = np.hstack((e_pairs[i][1].reshape(-1,1) for i in range(k)))
```

```python
W
```

```python
X_lda = X.dot(W)
X_lda
```

```python
plt.figure(figsize=(10,8))
colors=['b','g','r']
markers=['o','x','s']
for l,c,m in zip(np.unique(y),colors,markers):
  plt.scatter(X_lda[y == l, 0],X_lda[y == l,1],c=c,label=l,marker=m)

plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title('LDA Projection')
plt.legend()
plt.show()
```
